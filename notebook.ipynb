# Multiple categorical variables using stacked and grouped bar plots

# How are the cars in the data set distributed over doornumber and carbody?

```python
df.info()
```

```python
doornumber_carbody_crosstable = pd.crosstab(df["doornumber"], df["carbody"])
doornumber_carbody_crosstable
```

```python
carbody_categories = doornumber_carbody_crosstable.columns
carbody_categories
```

```python
fig ,ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc["four"] , alpha = 0.5)
bar2 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc["two"] , alpha = 0.5)
ax.set_xlabel("Index of categories")
ax.set_ylabel("count")
ax.set_title("Distribution of no. of cars over no. of doors and type of car body");
```

```python
fig ,ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc["four"] , width= 0.5)
bar2 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc["two"] , width= 0.5)
ax.set_xlabel("Index of categories")
ax.set_ylabel("count")
ax.set_title("Distribution of no. of cars over no. of doors and type of car body");
```

```python
fig ,ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.25 , doornumber_carbody_crosstable.loc["four"] , width= 0.5) #for orange one
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.25 , doornumber_carbody_crosstable.loc["two"] , width= 0.5) # for blue one
ax.set_xlabel("Index of categories")
ax.set_ylabel("count")
ax.set_title("Distribution of no. of cars over no. of doors and type of car body");
# Shifting to the left -0.25
# Shifting to the right +0.25
```

```python
fig ,ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.2 , doornumber_carbody_crosstable.loc["four"] , width= 0.4) #for orange one
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.2 , doornumber_carbody_crosstable.loc["two"] , width= 0.4) # for blue one
ax.set_xlabel("Index of categories")
ax.set_ylabel("count")
ax.set_title("Distribution of no. of cars over no. of doors and type of car body");
```

```python
fig ,ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.2 , doornumber_carbody_crosstable.loc["four"] , width= 0.4) #for orange one
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.2 , doornumber_carbody_crosstable.loc["two"] , width= 0.4) # for blue one
ax.set_xlabel("Index of categories")
ax.set_ylabel("count")
ax.set_title("Distribution of no. of cars over no. of doors and type of car body")
ax.set_xticks(np.arange(len(carbody_categories)))
ax.set_xticklabels(carbody_categories);
```

```python
fig ,ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.2 , doornumber_carbody_crosstable.loc["four"] , width= 0.4 , label = "four") #for orange one
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.2 , doornumber_carbody_crosstable.loc["two"] , width= 0.4 , label = "two") # for blue one
ax.set_xlabel("Index of categories")
ax.set_ylabel("count")
ax.set_title("Distribution of no. of cars over no. of doors and type of car body")
ax.set_xticks(np.arange(len(carbody_categories)))
ax.set_xticklabels(carbody_categories)
ax.legend();
```

# Draw different histograms for two door and four door cars

```python
two_door_df = df[df["doornumber"] == "two"]
four_door_df = df[df["doornumber"] == "four"]
```

```python
plt.hist(two_door_df["price"])
plt.hist(four_door_df["price"])
plt.xlabel("price")
plt.ylabel("number of cars")
plt.title("Distribution of prices of cars based on number of doors");
```

```python
plt.hist(two_door_df["price"] , alpha = 0.5 , label = "two")
plt.hist(four_door_df["price"] , alpha = 0.5 , label = "four")
plt.xlabel("price")
plt.ylabel("number of cars")
plt.title("Distribution of prices of cars based on number of doors")
plt.legend();
```

# Stacked histogram

# Representing the same info. is stacked histogram

```python
plt.hist([two_door_df["price"] , four_door_df["price"]] , alpha = 0.5 , label = ["two door" , "four door"] ,stacked = True)
plt.xlabel("price")
plt.ylabel("number of cars")
plt.title("Distribution of prices of cars based on number of doors")
plt.legend();
```

```python
fig, axs = plt.subplots(1, 2, figsize = (8,4))
```

```python
fig, axs = plt.subplots(1, 2, figsize = (8,4))

axs[0].hist(two_door_df["price"] , bins =20 , color="blue")
axs[0].set_title("Histogram of prices of two door cars")
axs[0].set_xlabel("price")
axs[0].set_ylabel("frequency")
axs[0].tick_params(axis = "x", rotation = 45) # parameters

axs[1].hist(four_door_df['price'], bins=20, color ="green")
axs[1].set_title("Histogram of prices of four door cars")
axs[1].set_xlabel("price")
axs[1].set_ylabel("frequency")
axs[1].tick_params(axis = "x", rotation = 45)

plt.suptitle("Histogram of prices of two door and four door cars")
plt.tight_layout();
```
